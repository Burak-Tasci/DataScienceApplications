```python
y = maliciousornot_data.loc[:,"Type"].values
X = maliciousornot_data.drop("Type", axis = 1).values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(units=30, activation = "relu"))
model.add(Dense(units=15, activation = "relu"))
model.add(Dense(units=15, activation = "relu"))
model.add(Dense(units=1, activation = "sigmoid"))

model.compile(loss = "binary_crossentropy", optimizer = "adam")
```

```python
model.fit(x = X_train, y = y_train,
         batch_size = 200,
         epochs = 700,
         validation_data=(X_test,y_test))
```

```python
loss_data = pd.DataFrame(model.history.history)
loss_data
```

```python
loss_data.plot(kind="line")
```

```python
model = Sequential()

model.add(Dense(units=30, activation = "relu"))
model.add(Dense(units=15, activation = "relu"))
model.add(Dense(units=15, activation = "relu"))
model.add(Dense(units=1, activation = "sigmoid"))

model.compile(loss = "binary_crossentropy", optimizer = "adam")
```

```python
early_stopping = EarlyStopping(mode="min",
                               monitor="val_loss",
                               patience = 25,
                            verbose = 1)
```

```python
model.fit(x = X_train, y = y_train,
         batch_size = 200,
         epochs = 700,
         callbacks = [early_stopping],
         validation_data=(X_test,y_test))
```

```python
loss_data = pd.DataFrame(model.history.history)
loss_data
```

```python
loss_data.plot(kind="line")
```

```python
model = Sequential()

model.add(Dense(units=30, activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(units=15, activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(units=15, activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(units=1, activation = "sigmoid"))

model.compile(loss = "binary_crossentropy", optimizer = "adam")
```

```python
model.fit(x = X_train, y = y_train,
         batch_size = 200,
         epochs = 700,
         callbacks = [early_stopping],
         validation_data=(X_test,y_test))
```

```python
loss_data = pd.DataFrame(model.history.history)
loss_data
```

```python
loss_data.plot(kind="line")
```

```python
preds = model.predict_classes(X_test, verbose = 1)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, preds)
cm
```

```python
np.sum(cm.diagonal()) / np.sum(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, preds))
```
